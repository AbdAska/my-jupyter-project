```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

# --- Custom tick formatter for the x-axis ---
def custom_log_format(x, pos):
    if np.isclose(x, 0.1, atol=1e-12):
        return "0"
    if np.isclose(x, round(x), atol=1e-6):
        return f"{int(round(x))}"
    return f"{x:g}"

# --- Define columns and compute the global maximum for y-axis limits ---
uniformity_columns = [
    'temp-uniformity-1', 
    'temp-uniformity-2', 
    'temp-uniformity-3', 
    'temp-uniformity-4', 
    'temp-uniformity-5', 
    'temp-uniformity-6'
]

max_z2 = my_filez2[uniformity_columns].max().max()
max_z1 = my_filez1[uniformity_columns].max().max()
global_ymax = max(max_z2, max_z1)

# --- Set global font parameters ---
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 20
plt.rcParams['font.weight'] = 'normal'

# --- Define colors for the two datasets ---
color_filez1 = '#1f77b4'
color_filez2 = '#ff7f0e'

# --- Define custom zone texts for each subplot ---
zone_texts = [
    "Zone-vO", "Zone-vC", "Zone-vI",
    "Zone-hO", "Zone-hC", "Zone-hI"
]

# --- Helper function to plot lines with solid and dotted transitions and shading ---
def plot_transition_line(ax, x_data, y_data, area_q, color, label, 
                         linestyle_solid='solid', linestyle_dotted='--', linewidth=2.5):
    # Find the transition point where area-q becomes 1
    transition_idx = np.where(area_q == 1)[0]
    if len(transition_idx) > 0:
        transition_idx = transition_idx[0]
        # Shade the area under the curve up to the transition point
        ax.fill_between(x_data[:transition_idx+1], y_data[:transition_idx+1], 0, 
                        color=color, alpha=0.3, zorder=1)
        # Plot the first part with solid line
        sns.lineplot(ax=ax, x=x_data[:transition_idx], y=y_data[:transition_idx],
                     label=label, linewidth=linewidth, color=color, linestyle=linestyle_solid)
        # Plot the second part with dotted line
        sns.lineplot(ax=ax, x=x_data[transition_idx:], y=y_data[transition_idx:],
                     label=label, linewidth=linewidth, color=color, linestyle=linestyle_dotted)
    else:
        # Shade the entire area if no transition happens
        ax.fill_between(x_data, y_data, 0, color=color, alpha=0.3, zorder=1)
        # Plot as solid line
        sns.lineplot(ax=ax, x=x_data, y=y_data,
                     label=label, linewidth=linewidth, color=color, linestyle=linestyle_solid)

# --- Create a 2-row x 3-column figure ---
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(10, 7.5))
# Flatten the axes for easier indexing
axs = axs.flatten()
nrows = 2
ncols = 3

for i, col in enumerate(uniformity_columns):
    ax = axs[i]
    
    # Plot data from both files using the transition function with shading.
    plot_transition_line(ax, my_filez2['flow-time'], my_filez2[col], my_filez2['area-q'], color_filez2, 'File z2')
    plot_transition_line(ax, my_filez1['flow-time'], my_filez1[col], my_filez1['area-q'], color_filez1, 'File z1')
    
    # Remove any automatically generated legend.
    leg = ax.get_legend()
    if leg:
        leg.remove()
    
    # Set logarithmic scale and x-axis limits.
    ax.set_xscale('log')
    ax.set_xlim(1, 4500)

    # Determine the row (top: row_idx == 0, bottom: row_idx == 1).
    row_idx = i // ncols
    if row_idx == 1:
        # Bottom row: configure major ticks and custom formatter.
        ax.xaxis.set_major_locator(ticker.FixedLocator([1, 10, 100, 1000]))
        ax.xaxis.set_major_formatter(ticker.FuncFormatter(custom_log_format))
        ax.xaxis.set_minor_locator(ticker.LogLocator(base=10, subs=np.arange(1, 10), numticks=100))
        ax.xaxis.set_minor_formatter(ticker.NullFormatter())
        # Force ticks only at the bottom.
        ax.xaxis.set_ticks_position('bottom')
        ax.tick_params(axis='x', which='major', bottom=True, top=False, labelbottom=True)
        ax.tick_params(axis='x', which='minor', bottom=True, top=False)
        ax.set_xlabel('log(t) (s)')
        # Remove any automatic offset text (e.g., "1e0").
        ax.xaxis.get_offset_text().set_visible(False)
    else:
        # Top row: remove all x-axis ticks and labels.
        ax.set_xticks([]) 
        ax.set_xticks([], minor=True) 
        ax.set_xlabel('')
    
    # Set uniform y-axis limits.
    ax.set_ylim(0, 14.5)
    
    # Standardize tick appearance.
    ax.tick_params(axis='both', which='major', length=5, width=1)
    
    # Remove y-axis ticks/labels for subplots not in the first column.
    col_idx = i % ncols
    if col_idx != 0:
        ax.set_ylabel('')
        ax.set_yticklabels([])
        ax.tick_params(axis='y', which='both', left=False, right=False)
    else:
        ax.set_ylabel(r'$\mathregular{{U}_{T}}$ (°C)')
    
    # --- Add an editable text box with custom text ---
    text_to_show = zone_texts[i] if i < len(zone_texts) else "zone"
    ax.text(0.05, 0.955, text_to_show, transform=ax.transAxes,
            fontsize=20, fontweight='normal', verticalalignment='top',
            bbox=dict(facecolor='lightyellow', edgecolor='black', boxstyle='round,pad=0.1'))

# Adjust spacing between subplots.
plt.subplots_adjust(wspace=0.02, hspace=0.02)

# Save and display the figure.
plt.savefig('66.png', dpi=400, bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np

# ── User-tunable parameters ─────────────────────────────────────────────────
wspace               = 0.25                # horizontal space between ax1 and ax2
right_axis_color     = 'red'               # color for the right y-axis on ax2
ax1_label_coords     = (-0.12, 0.5)        # (x, y) for ax1’s y-label in axes coords
ax2_left_label_coords= (-0.17, 0.5)        # (x, y) for ax2’s left y-label in axes coords

# --- Custom tick formatter for ax1’s x-axis ---
def custom_log_format(x, pos):
    return "0" if np.isclose(x, 0.1, atol=1e-12) else f"{x:g}"

# --- Helper: solid-to-dotted transition lines with shading (ax1) ---
def plot_transition_line(ax, x_data, y_data, area_q, color, label,
                         linestyle_solid='solid', linestyle_dotted='--', linewidth=2.5):
    idx = np.where(area_q == 1)[0]
    if idx.size:
        i = idx[0]
        ax.fill_between(x_data[:i+1], y_data[:i+1], 0, color=color, alpha=0.3, zorder=1)
        sns.lineplot(ax=ax, x=x_data[:i], y=y_data[:i],
                     label=label, linewidth=linewidth,
                     color=color, linestyle=linestyle_solid)
        sns.lineplot(ax=ax, x=x_data[i:], y=y_data[i:],
                     linewidth=linewidth, color=color,
                     linestyle=linestyle_dotted)
    else:
        ax.fill_between(x_data, y_data, 0, color=color, alpha=0.3, zorder=1)
        sns.lineplot(ax=ax, x=x_data, y=y_data,
                     label=label, linewidth=linewidth,
                     color=color, linestyle=linestyle_solid)

# ── global font ─────────────────────────────────────────────────────────────
plt.rcParams.update({
    'font.family': 'Times New Roman',
    'font.size'  : 22,
    'font.weight': 'normal'
})

# ── create figure + two subplots with width ratio & wspace ─────────────────
fig, (ax1, ax2) = plt.subplots(
    1, 2, figsize=(12, 7),
    gridspec_kw={'width_ratios': [0.6, 0.4], 'wspace': wspace}
)

############################################
# LEFT SUBPLOT (ax1): log-line + inset (unchanged)
############################################
plot_transition_line(ax1, my_filez1['flow-time'], my_filez1['heat-storage-rate'],
                     my_filez1['area-q'], '#2178b4', 'case-1')
plot_transition_line(ax1, my_filez2['flow-time'], my_filez2['heat-storage-rate'],
                     my_filez2['area-q'], '#ff861b', 'case-2')

ax1.set_xscale('log')
ax1.set_xlim(0.1, 5000)
ax1.xaxis.set_major_formatter(FuncFormatter(custom_log_format))
ax1.set_xlabel(r'log(t) (s)')
ax1.set_ylabel(r'$\mathregular{Q_{s}}$ (W)')
# apply the user-controlled label coords:
ax1.yaxis.set_label_coords(*ax1_label_coords)
ax1.set_ylim(0, 1000)
ax1.tick_params(axis='both', length=6, width=1)
if ax1.get_legend():
    ax1.get_legend().remove()

# inset bar inside ax1
rect = ax1.get_position()
inset_left   = rect.x0 + 0.625 * rect.width
inset_bottom = rect.y0 + 0.491 * rect.height
inset_width  = 0.365 * rect.width
inset_height = 0.5   * rect.height

ax_inset = fig.add_axes([inset_left, inset_bottom, inset_width, inset_height])
bars = ax_inset.bar(my_file0['model'], my_file0['flux'], width=0.75,
                    color=['#2178b4', '#ff861b'], zorder=2)
for bar, val in zip(bars, my_file0['flux']):
    ax_inset.text(bar.get_x() + bar.get_width()/2, val + 0.1,
                  f'{val:.2f}', ha='center', va='bottom', fontsize=18)

ax_inset.set_ylabel(r'$\mathregular{\overline{Q}_{s}}$ (W)', fontsize=18, labelpad=0)
ax_inset.yaxis.set_label_coords(-0.30, 0.62)
ax_inset.set_xlim(-0.5, len(my_file0['model'])-0.5)
ax_inset.set_ylim(5, 350)
ax_inset.set_xticks([])
ax_inset.tick_params(axis='y', length=6, labelsize=20)

############################################
# RIGHT SUBPLOT (ax2): bars + twin axis (no legend)
############################################
x = np.arange(len(my_file0['model']))
bars2 = ax2.bar(x, my_file0['heat-storage-density'],
                width=0.4, color=['#2178b4', '#ff861b'], zorder=2)
for bar, pct in zip(bars2, my_file0['percentage change density']):
    h = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, h,     f'{h:.2f}',
             ha='center', va='bottom', fontsize=20, zorder=5)
    ax2.text(bar.get_x() + bar.get_width()/2, h*0.95,
             (f'+{pct:.1f}%' if pct>=0 else f'{pct:.1f}%'),
             ha='center', va='top', fontsize=20, color='white', zorder=5)

ax2.set_ylabel(r'$\mathregular{S_{d}}$ (KJ/kg)')
# apply user-controlled coords for ax2’s left y-label:
ax2.yaxis.set_label_coords(*ax2_left_label_coords)
ax2.set_ylim(0, 350)
ax2.tick_params(axis='both', length=6, width=1)
ax2.set_xticks(x)
ax2.set_xticklabels(my_file0['model'], rotation=0)

# ── add right-hand twin axis (no legend) ────────────────────────────────────
axR = ax2.twinx()
axR.plot(x, my_file0['khatamata'],
         marker='o', linestyle='-', linewidth=2.5,
         color=right_axis_color)
axR.set_ylabel(r'$\mathregular{S_{R}}$ (KJ/kg·s)', rotation=90, labelpad=2)
axR.tick_params(axis='y', length=6, width=1, colors=right_axis_color)
axR.yaxis.label.set_color(right_axis_color)
axR.spines['right'].set_color(right_axis_color)
for spine in ['top','left','bottom']:
    axR.spines[spine].set_visible(False)

############################################
# Final export
############################################
plt.savefig('combined_plot.png', dpi=400, bbox_inches='tight')
plt.show()
```
