```python
import matplotlib.pyplot as plt
import numpy as np

# Set up figure and axis
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_aspect("equal")
ax.axis("off")

# Define radii for each level
center_radius = 0.1
level1_radius = 0.2
level2_radius = 0.3
level3_radius = 0.35
level4_radius = 0.95

# Define colors for each level
colors = ['#7f00ff', '#32cd32', '#00bfff', '#4169e1']

# Center text
center_text = "Water-\nHarvesting\nMaterial"
ax.text(0.5, 0.5, center_text, ha='center', va='center', transform=ax.transAxes, fontsize=14, color='white', weight='bold')

# Draw and label each level separately

# Level 1
level1_texts = ["Spider Silk", "Cactus Stem", "Desert Beetle", "Nepenthes Pitcher", "Namib Desert"]
for i, text in enumerate(level1_texts):
    angle_rad = np.deg2rad(i * 360 / len(level1_texts))
    x = 0.5 + level1_radius * np.cos(angle_rad) * 1.1
    y = 0.5 + level1_radius * np.sin(angle_rad) * 1.1
    rotation = np.rad2deg(angle_rad) + 90
    ax.text(x, y, text, ha='center', va='center', transform=ax.transAxes, fontsize=10, color='white', rotation=rotation, rotation_mode='anchor')

circle1 = plt.Circle((0.5, 0.5), level1_radius, color=colors[0], transform=ax.transAxes, fill=True, alpha=0.6)
ax.add_patch(circle1)

# Level 2
level2_texts = ["Dip-coating", "Photo/mechanical lithography", "Fluid coating", "Magnetic particle-assisted molding"]
for i, text in enumerate(level2_texts):
    angle_rad = np.deg2rad(i * 360 / len(level2_texts))
    x = 0.5 + level2_radius * np.cos(angle_rad) * 1.2
    y = 0.5 + level2_radius * np.sin(angle_rad) * 1.2
    rotation = np.rad2deg(angle_rad) + 90
    ax.text(x, y, text, ha='center', va='center', transform=ax.transAxes, fontsize=10, color='white', rotation=rotation, rotation_mode='anchor')

circle2 = plt.Circle((0.5, 0.5), level2_radius, color=colors[1], transform=ax.transAxes, fill=True, alpha=0.5)
ax.add_patch(circle2)

# Level 3
level3_texts = ["Chemical or electrochemical etching", "Electro hydrodynamic", "Thermal bonding", "Microfluidic pumping"]
for i, text in enumerate(level3_texts):
    angle_rad = np.deg2rad(i * 360 / len(level3_texts))
    x = 0.5 + level3_radius * np.cos(angle_rad) * 1.25
    y = 0.5 + level3_radius * np.sin(angle_rad) * 1.25
    rotation = np.rad2deg(angle_rad) + 90
    ax.text(x, y, text, ha='center', va='center', transform=ax.transAxes, fontsize=10, color='white', rotation=rotation, rotation_mode='anchor')

circle3 = plt.Circle((0.5, 0.5), level3_radius, color=colors[2], transform=ax.transAxes, fill=True, alpha=0.5)
ax.add_patch(circle3)

# Level 4
level4_texts = ["Environmental patterning", "Pumping technology", "Advanced materials", "Biomechanical design"]
for i, text in enumerate(level4_texts):
    angle_rad = np.deg2rad(i * 360 / len(level4_texts))
    x = 0.5 + level4_radius * np.cos(angle_rad) * 1.3
    y = 0.5 + level4_radius * np.sin(angle_rad) * 1.3
    rotation = np.rad2deg(angle_rad) + 90
    ax.text(x, y, text, ha='center', va='center', transform=ax.transAxes, fontsize=10, color='white', rotation=rotation, rotation_mode='anchor')

circle4 = plt.Circle((0.5, 0.5), level4_radius, color=colors[3], transform=ax.transAxes, fill=True, alpha=0.4)
ax.add_patch(circle4)

plt.show()
```
